```python
import numpy as np
import pandas as pd
```

```python
regioni_info = pd.read_csv('../data/italy_regions.csv', sep=';', usecols=['regione','superficie','num_comuni', 'num_provincie'])
regioni_info
```

```python
regioni_istat = pd.read_csv('../data/popolazioneISTAT-regioni1gen2020.csv', sep=';')
regioni_istat
```

```python
regioni_pop_info=pd.merge(regioni_istat, regioni_info, left_on="Territorio", right_on="regione").drop('regione', axis=1)
regioni_pop_info
```

Operazione simile alla **gather()** di R: vgliamo trasformare il data frame in modo da avere Maschi, Femmine e Totale come valori di una colonna chiave. In Phython c'è la funzione **melt()** che fa una cosa simile, non identica. Qui c'è una buona discussione: https://medium.com/yulun-ding-ambassador/a-beginner-guides-and-example-of-shifting-between-r-and-python-pandas-b7039efd8c75

```python
regioni_pop_chiave=regioni_pop_info.melt(
               id_vars=["Territorio","superficie","num_comuni", "num_provincie"],  
               var_name="Sesso",             
               value_name="Popolazione"            
               )
regioni_pop_chiave
```

Ora vediamo alcuni esempi per capire bene la differenza tra stack() e unstack(). Sulla documentazione di pandas trovate una buona sintesi con esempiosemplice: https://pandas.pydata.org/pandas-docs/stable/user_guide/reshaping.html

```python
regioni_pop_chiave2 = regioni_pop_chiave.set_index('Territorio')
regioni_pop_chiave2
```

Con **stack()** si opera su un dataframe indicizzato e questo trasforma l'indice in colonna (o più colonne se abbiamo un indice a più livelli) con i valori delle colonne originali impilati come celle.

```python
regioni_pop_chiave2.stack()
```

```python
regioni_pop_chiave3=regioni_pop_chiave.set_index(['Territorio','Sesso'])
regioni_pop_chiave3
```

```python
regioni4=regioni_pop_chiave3.stack()
regioni4
```

Con il dataframe indicizzato posso usare la **loc()** per selezionare per indice. Vedere anche https://towardsdatascience.com/effective-data-filtering-in-pandas-using-loc-40eb815455b6

```python
regioni4.loc[['Toscana', 'Piemonte']]
```

```python
regioni4.loc['Toscana','Maschi']
```

Con **unstack()** si fa l'operazione opposta. Con un multi indice, **unstack(0)** riporta a colonne i valori dell'indice più interno (Sesso), con **unstack(1)** si sale di livello (Territorio)

```python
regioni_pop_chiave3.unstack(1)
```

```python
regioni_pop_chiave3.unstack(0)
```

Infine, se ho un data frame non indicizzato, per la selezione delle righe o uso la **iloc()** che sfrutta l'indice implicito di righe e colonne oppure la **query()** (analoga alla select() di R)

```python
regioni_pop_chiave3
regioni_pop_chiave3.query('num_comuni>1000')
```
